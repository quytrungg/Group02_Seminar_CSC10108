```python
avg_data = df[df["Gender"].isin(["Man", "Woman"])]
avg_data = avg_data[["Country", "Gender", "Salary", "Year"]]
avg_data = avg_data.dropna(subset=["Gender", "Country", "Salary", "Year"])
avg_data = avg_data.groupby(["Country", "Gender", "Year"], as_index=False).mean()


fig = px.bar(avg_data,
            x="Country", y="Salary",
            color="Gender",
            animation_frame="Year",
            barmode="group",
            width=_WIDTH, height=600,
            category_orders={"Country": ["ChinağŸ‡¨ğŸ‡³", "IndiağŸ‡®ğŸ‡³", "Viet Nam", "U.S.ğŸ‡ºğŸ‡¸"],
                            "Gender":["Man", "Woman"],
                            "Year": range(2018,2022)},
            title="Average yearly compensation of men and women",
            text="Salary")

fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(yaxis_range=[0,38000])
fig.update_yaxes(title="Average yearly compensation (number of Big Macs)")
fig.update_xaxes(type='category')
fig.show()
```

```python
fig = make_subplots(rows=1, cols=2)

top_10_languages = df['Languages'].value_counts().nlargest(10)
top_10_ides = df['IDEs'].value_counts().nlargest(10)

fig.add_trace(
    go.Bar(y=top_10_languages.index,
           x=top_10_languages.values,
           orientation='h',
           text=top_10_languages,
           textposition="auto",
           name="Top 10 Languages"),
    row=1, col=1
)

fig.add_trace(
    go.Bar(y=top_10_ides.index,
           x=top_10_ides.values,
           orientation='h',
           text=top_10_ides,
           textposition="auto",
           name="Top 10 IDEs"),
    row=1, col=2
)

fig.update_layout(height=600, width=1800, title_text="Languages and IDEs")
fig.show()
```
