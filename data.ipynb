```python
import pandas as pd
import numpy as np
```

```python
choice = pd.read_csv('./local_data/multiple_choice_responses.csv').fillna('')
question =  pd.read_csv('./local_data/questions_only.csv').fillna('')
text_reponse = pd.read_csv('./local_data/other_text_responses.csv').fillna('')
survey_s = pd.read_csv('./local_data/survey_schema.csv').fillna('')
```

```python
multichoice =  set(map(lambda x: x.split('_')[0] ,list(filter( lambda x : 'Part' in x ,choice.columns))))
```

```python
Q = {}
for i in list(filter( lambda x: x.split('_')[0] in multichoice ,choice.columns)):
    if len(i.split('_')) != 1:
        try:
            Q[i.split('_')[0]][i.split('_')[-1]]= choice.iloc[0][i]
        except:
            Q[i.split('_')[0]] =  {i.split('_')[-1]: choice.iloc[0][i]}
    else:
        try:
            Q[i.split('_')[0]]['ANSWER']= choice.iloc[0][i]
        except:
            Q[i.split('_')[0]]= {'ANSWER': choice.iloc[0][i] }
for k,v in Q.items():
    if len (Q[k]) > 1:
        for k_item,v_item in Q[k].items():
            Q[k][k_item] = v_item.split(' - ')[-1].strip()
```

```python
for i in Q.keys():
    col = list(filter(lambda x: i in x and 'Part' in x, choice.columns.to_list()))
    choice[i] = choice[col].apply(lambda row:[ i+1  for i in range(len(row)) if row[i] !='' ], axis=1)
    choice.drop(col, axis='columns', inplace=True)
```

```python
col = list(filter(lambda x:'OTHER' in x, choice.columns.to_list()))
choice.drop(col, axis='columns', inplace=True)
choice.drop(index = 0, inplace = True)
choice
```
